```python
!pip install langchain
!pip install langchain_community
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline

# Create pipeline
pipe = pipeline(
    "text-generation",
    model="gpt2"
)

# Wrap with LangChain
llm = HuggingFacePipeline(pipeline=pipe)

# Run query
response = llm.invoke("Explain LangChain in simple words")

print(response)
```

```python
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="gpt2",
    max_new_tokens=100,
    temperature=0.7
)

llm = HuggingFacePipeline(pipeline=pipe)

response = llm.invoke(
    "Explain LangChain in simple words"
)

print(response)
```

```python
# Prompt templates

from langchain_core.prompts import PromptTemplate

# Create template
template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms in 3 sentences only."
)

# Format prompt
formatted_prompt = template.format(topic="Computer Vision")

# Pass to LLM
response = llm.invoke(formatted_prompt)

print(response)
```

```python
# Chains

from langchain_classic.chains import LLMChain

# Reuse previous template
template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms in 3 sentences only."
)
chain = LLMChain(
    llm=llm,
    prompt=template
)

# Run chain
response = chain.invoke("Artificial Intelligence")

print(response)
```

```python
# Memory

from langchain_classic.memory import ConversationBufferMemory
from langchain_classic.chains import ConversationChain

# Initialize memory
memory = ConversationBufferMemory()

# Create conversation chain
conversation = ConversationChain(
    llm=llm,
    memory=memory
)

# First interaction
print(conversation.predict(input="Remember: My name is Pavan"))
# Second interaction (should remember name)
print(conversation.predict(input="What is my name?"))
```

```python
# Agents and tools

from langchain_classic.agents import initialize_agent, Tool

# Create a simple calculator tool
def calculator(query):
    return str(eval(query))

tool = Tool(
    name="Calculator",
    func=calculator,
    description="Useful for solving math problems"
)

# Initialize agent
agent = initialize_agent(
    tools=[tool],
    llm=llm,
    agent="zero-shot-react-description",
    verbose=True,
    handle_parsing_errors=True   # Important Fix right here
)

response = agent.run("Calculate 25 * 4?")
print(response)
```
