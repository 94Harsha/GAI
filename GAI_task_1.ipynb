```python
from langchain_core.prompts import PromptTemplate

# Step 1: Create Prompt Template
template = "Explain {topic} in simple terms for beginners"

prompt = PromptTemplate(
    input_variables=["topic"],
    template=template
)

# Step 2: Use the template
output = prompt.format(topic="Artificial Intelligence")

print(output)
```

```python
from langchain_core.prompts import PromptTemplate

# Step 1: Create template
template = "Explain {topic} for {audience} in a {tone} tone"

prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# Step 2: Test cases
output1 = prompt.format(topic="AI", audience="beginners", tone="friendly")
output2 = prompt.format(topic="Python", audience="kids", tone="fun")
output3 = prompt.format(topic="Deep Learning", audience="engineers", tone="technical")

print(output1)
print(output2)
print(output3)
```

```python
from langchain_core.prompts import PromptTemplate

# Topic
topic = "Machine Learning"

# Teaching Template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Storytelling Template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Generate outputs
teaching_output = teaching_template.format(topic=topic)
interview_output = interview_template.format(topic=topic)
story_output = story_template.format(topic=topic)

# Print results
print("Teaching Prompt:")
print(teaching_output)

print("\nInterview Prompt:")
print(interview_output)

print("\nStorytelling Prompt:")
print(story_output)
```
